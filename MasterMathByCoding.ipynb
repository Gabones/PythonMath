```python
import sympy as sym
import math

def erroRel(x1, x2):
  return x1 if x2 == 0 else ((x1-x2)/x2)

def zeroBissec(f, a, b, tol, imax):
  i = 1
  p = 0.5*(a+b)
  fa = f.subs(x,a)
  fb = f.subs(x,b)
  fp = f.subs(x,p)
  erel = math.inf
  while(i < imax and abs(erel) > tol):
    if fa*fp <= 0:
      b = p
      fb = f.subs(x,p)
    else:
      a = p
      fa = f.subs(x,p)
    i += 1
    u = p
    p = 0.5*(a+b)
    fp = f.subs(x,p)
    erel = erroRel(u,p)
  return [p,erel,i]

x = sym.symbols('x')

fx = x**2 - 16

print(zeroBissec(fx,0,10,0.001,100))

```

###Aula 126 - Graphing tangent lines: exercise

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

```python
def computeTangent(f, xa, bnds):
  #defining the derivative and value at point xa
  df = sym.diff(f)
  fa = f.subs(x,xa)
  dfa = df.subs(x,xa)

  #evaluate the tangent line 
  xx = np.linspace(bnds[0], bnds[1], 200)

  return dfa * (xx - xa) + fa
```

```python
x = sym.symbols('x')
f = x**2
xx = np.linspace(-2, 2, 200)
f_lamb = sym.lambdify(x, f)(xx)

for a in xx:
  y = computeTangent(f,a,xx[[0,-1]])
  plt.plot(xx,y, color=[ abs(a)/3, abs(a)/4, abs(a)/2])

plt.plot(xx, f_lamb)
plt.axis('square')
plt.axis([-2,2,-1,3])
plt.axis('off')
plt.show()
```

####Computing the area between two functions

The area between two functions is given by the formula $A = \int_a^b f(x) - g(x)dx$

In our example, $f(x) = x^2$ and $g(x) = x$

Therefore, $A = \int_a^b (x^2 - x)dx$

We will compute the area between the two crossing points, that is, where the two functions are equal. This is given by the two solutions to the equation $x^2 = x$

The two solutions are $x = 0$ and $x = 1$. This gives us the definite integral of $A = \int_0^1 (x^2 - x) dx$

The final answer will be given in the code below.

```python
p_intersect = sym.solve(symf - symg)
display(Math('f(x) \\text{ and } g(x) \\text{ intersect at } x = %g \\text{ and } x = %g' %(p_intersect[0], p_intersect[1])))

# Outro jeito de calcular
#A = sym.integrate(symg, (x, p_intersect[0], p_intersect[1])) - sym.integrate(symf, (x, p_intersect[0], p_intersect[1]))
A = sym.integrate(symg - symf, (x, p_intersect[0], p_intersect[1]))
print()
display(Math('\\text{ the area between these two functions is } %s' %A))
```

```python
#Exercise

def MultiplyMatrix(A, B):
  if(np.shape(A)[1] != np.shape(B)[0]):
    raise ValueError("The inner dimensions dont match")
  
  result = np.zeros((np.shape(A)[0], np.shape(B)[1]))

  for l in range(np.shape(A)[0]):
    for c in range(np.shape(B)[1]):
      result[l, c] = np.dot(A[l, :], B[:, c])

  return result

print(MultiplyMatrix(C, B))
print()
print(C @ B)
```

###Aula 148 - Autovalores e Autovetores
Um interpreta√ß√£o alg√©brica e geom√©trica de autovalores e autovetores

Como executar uma autodecomposi√ß√£o em python: np.linalg.eig()

O que significa diagonaliza√ß√£o?

A eigendecomposition ou autodecomposi√ß√£o de uma matriz s√≥ √© definida para matrizes quadradas, para matrizes n√£o quadradas existe uma decomposi√ß√£o em valores singulares que ser√° abordada futuramente. O resultado da autodecomposi√ß√£o de uma matrix MxM s√£o M autovalores e M autovetores.

Os valores do autovetor podem ser transformados em uma matriz diagonal tornando todo o processo uma multiplica√ß√£o de matrizes.


```python
M = np.random.randint(-5,6, [5,5])
M = M@M.T #garante que teremos uma autodecomposi√ß√£o em valores reais e n√£o em valores complexos (pesquisar)

eigvals, eigvecs = np.linalg.eig(M)

fig, ax = plt.subplots(1,3)

ax[0].imshow(M)
ax[0].set_title('Matriz')

ax[1].imshow(np.diag(eigvals))
ax[1].set_title('Autovalores diag')

ax[2].imshow(M)
ax[2].set_title('Autovetores')

plt.show()

Mv = M @ eigvecs[:,0] #multiplicar uma matriz por um vetor
lv = eigvals[0] * eigvecs[:,0] #tem o mesmo resultado de multiplicar por um escalar

print(f'{Mv}\n{lv}')
```
